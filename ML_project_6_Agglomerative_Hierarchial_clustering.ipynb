```python
import pandas as pd
df = pd.read_csv("/content/Mall_Customers_2.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df[['Annual Income (k$)','Spending Score (1-100)']]
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=15,10
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.axhline(y=8,color='r',linestyle='--')
plt.title('Dendogram')
plt.xlabel('Customer')
plt.ylabel('Euclidean')
plt.show()
```

the number of cluster will be the number of vertical line is inersected by the red line , no of cluster : 5

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
model.fit(x)
```

```python
y = model.fit_predict(x)
y
```

```python
import seaborn as sns
plt.rcParams['figure.figsize']=10,5
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue=y,data=df,s=40,palette=['green','orange','brown','dodgerblue','red'])
plt.title('Agglomerative Hierarchial clustering')
```
