```python
import pandas as pd
df = pd.read_csv('/content/student-mat.csv')
df
```

```python
df['G_Average'] = (df['G1'] + df['G2'] + df['G3']) / 3
```

```python
df.head()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['address','G1','G2','G3','famsize','Walc','Dalc','Medu','Fedu','Mjob','Fjob','reason','traveltime','schoolsup','nursery','famrel','freetime','romantic'],axis=1)
```

```python
df.info()
```

```python
df
```

```python
df['famsup'].value_counts()
```

```python
df['school'] = df['school'].map({'GP':0,'MS':1})
df['sex'] = df['sex'].map({'F':0,'M':1})
df['Pstatus'] = df['Pstatus'].map({'T':0,'A':1})
df['guardian'] = df['guardian'].map({'mother':0,'father':1,'other':2})
df['famsup'] = df['famsup'].map({'no':0,'yes':1})
df['paid'] = df['paid'].map({'no':0,'yes':1})
df['activities'] = df['activities'].map({'no':0,'yes':1})
df['higher'] = df['higher'].map({'no':0,'yes':1})
df['internet'] = df['internet'].map({'no':0,'yes':1})
```

```python
df
```

```python
y = df['G_Average']
x = df.drop(['G_Average'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import pandas as pd
df = pd.read_csv('/content/diabetes.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
y = df['Outcome']
x = df.drop(['Outcome'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
neighbor=np.arange(1,11)
train_accuracy=[]
test_accuracy=[]
for i in neighbor:
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(x_train,y_train)
  train_accuracy.append(knn.score(x_train,y_train))
  test_accuracy.append(knn.score(x_test,y_test))

print(train_accuracy)
print(test_accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbor,train_accuracy,label='train accuracy')
plt.plot(neighbor,test_accuracy,label='test accuracy')
plt.legend()
plt.xlabel('neighbor')
plt.ylabel('accuracy')
plt.show()
#
```
